# Read a .csv from a URL with Pandas
Link: root, season, league

```python
# Link structure
"https://football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
# Create root variable
root = "https://football-data.co.uk/mmz4281/"
```

### Multiple Leagues

-https://football-data.co.uk/mmz4281/2425/E0.csv

-https://football-data.co.uk/mmz4281/2425/E1.csv

-https://football-data.co.uk/mmz4281/2425/E2.csv

-https://football-data.co.uk/mmz4281/2425/E3.csv

-https://football-data.co.uk/mmz4281/2425/EC.csv

```python
# Creating the list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# Looping through leagues, read multiple csv and append them into a list
for league in leagues:
    df = pd.read_csv(root+ '2425' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd, and 3rd element
frames[1]
```

# Multiple Seasons

```python
# Creating the list of leagues
leagues = ['E0', 'E1', 'E2']
frames = []

# Looping through the range of seasons and insert a new column
for league in leagues:
    for season in range(20, 25):
        df = pd.read_csv(root + (str(season) + str(season+1)) + '/' + league + '.csv')
        df.insert(1, 'Season', season)
        frames.append(df)
```

```python
# Total frames (5 seasons x 3 leagues)
len(frames)
```

```python
# Show 1st, and last element (new season column was created)
frames[12]
```

# Organize All The Data in a Dictonary

```python
# Create a dictionary with the original name of the league as the key
dict_countries = {
    'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
    'German Bundesliga':'D1',
    'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3'
}
```

```python
# Get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}


# Looping through the range of seasons and insert a new column
for league in dict_countries:
    frames = []
    for season in range(20, 25):
        df = pd.read_csv(root + (str(season) + str(season+1)) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'Season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# Show dataframe inside dict_historical_data
dict_historical_data['English Premier League']
```
